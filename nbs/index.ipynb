## Developer Guide

If you are new to using `nbdev` here are some useful pointers to get you started.

### Install solveit_dmtools in Development mode

```sh
# make sure solveit_dmtools package is installed in development mode
$ pip install -e .

# make changes under nbs/ directory
# ...

# compile to have changes apply to solveit_dmtools
$ nbdev_prepare
```

### Installation

Install latest from the GitHub [repository][repo]:

```sh
$ pip install git+https://github.com/shuane/solveit_dmtools.git
```

or from [pypi][pypi]


```sh
$ pip install solveit_dmtools
```
